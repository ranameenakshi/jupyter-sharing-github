```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("/Users/meenakshirana/Desktop/new take off folder/data science study udemy/Countries_exercise.csv")
data
```

```python
x = data.iloc[:,1:]
x
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_cluster = data.copy()
data_with_cluster['Cluster'] = identified_clusters
```

```python
data_with_cluster
```

```python
plt.scatter(data_with_cluster['Longitude'],data_with_cluster['Latitude'],c = data_with_cluster['Cluster'],cmap = 'rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
