```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("/Users/meenakshirana/Desktop/new take off folder/data science study udemy/S38_L265/3.12. Example.csv")
data
```

```python
plt.scatter(data['Satisfaction'],data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyality')
plt.show()
```

```python
x = data.copy()
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
clusters['cluster_pred'] = kmeans.fit_predict(x)
```

```python
plt.scatter(clusters['Satisfaction'],clusters['Loyalty'],c = clusters['cluster_pred'],cmap = 'rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(i)
    kmeans = kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
wcss    
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans_new = KMeans(4)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new['Clusters_pred']  = kmeans_new.fit_predict(x_scaled)
clusters_new
```

```python
plt.scatter(clusters_new['Satisfaction'],clusters_new['Loyalty'],c = clusters_new['Clusters_pred'],cmap = 'rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```
