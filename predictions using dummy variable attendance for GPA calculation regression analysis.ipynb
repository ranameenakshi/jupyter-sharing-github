```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv("/Users/meenakshirana/Desktop/new take off folder/data science study udemy/S33_L204/1.03. Dummies.csv")
```

```python
data = raw_data.copy()
```

```python
data["Attendance"] = data["Attendance"].map({"Yes":1,"No":0})
```

```python
data
```

```python
data.describe()
```

```python
y = data["GPA"]
x1 = data[["SAT","Attendance"]]
```

```python
x = sm.add_constant(x1)
reasults = sm.OLS(y,x).fit()
reasults.summary()
```

```python
plt.scatter(data["SAT"],y)
yhat_no = 0.644 + 0.0014*data["SAT"]
yhat_yes = 0.867 + 0.0014*data["SAT"]
yhat = .275 + 0.0017*data["SAT"]
fig = plt.plot(data["SAT"],yhat_no, lw = 2, c = "red")
fig = plt.plot(data["SAT"],yhat_yes, lw = 2, c = "green")
fig = plt.plot(data["SAT"],yhat, lw = 4, c = "black")
plt.xlabel("SAT", fontsize = 20)
plt.ylabel("GPA",fontsize = 20)
plt.show()
```
